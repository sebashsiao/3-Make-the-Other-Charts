#### Setup

```python
# import matplotlib and pandas
from matplotlib import pyplot as plt
import pandas as pd

# load dataset
pie_data = pd.read_csv('EV_market_share.csv')
```

```python
# display charts correctly
%matplotlib inline
plt.rcParams['figure.figsize'] = (5, 3)
plt.rcParams['figure.dpi'] = 75
```

1. Run the Setup cells to load in the necessary packages and the `EV_market_share` csv. Run the cell below to view the columns in the `pie_data` dataset.

```python
pie_data.head()
```

2. Write the code to make a pie chart of the `‘number_sold’` column.

```python
## YOUR SOLUTION HERE ##
plt.pie(x=pie_data.number_sold)
plt.title('Top 10 EVs sold in the US by count, 2021')
plt.show()
```

3. Add labels to the pie chart using the `‘model’` column of the dataset.

```python
## YOUR SOLUTION HERE ##
plt.pie(pie_data.number_sold, labels=pie_data.model)
plt.title('Top 10 EVs sold in the US by count, 2021')


plt.show()
```

4. Set the start angle equal to 60.

```python
## YOUR SOLUTION HERE ##
plt.pie(pie_data.number_sold, labels=pie_data.model, startangle=60)
plt.title('Top 10 EVs sold in the US by count, 2021')


plt.show()
```

<details>
    <summary style="display:list-item; font-size:16px; color:blue;"><i>What can we take away from this pie chart? Toggle to check:</i></summary>
    This pie chart is a great demonstration of Tesla's market dominance in the US. Viewers are able to see that two Tesla models account for about two thirds of the total count of EVs sold in the US in 2021. The Tesla Model Y accounts for almost half the sales count, and the Tesla Model 3 accounts for about a quarter of the sales count.<br><br>
